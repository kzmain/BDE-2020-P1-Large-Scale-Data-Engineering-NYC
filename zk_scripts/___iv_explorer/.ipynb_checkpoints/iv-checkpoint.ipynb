```python
import os
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Set up modes and dirs
databricks = False
overwrite  = False
is_yellow  = False
yellow     = "yellow" if is_yellow else "foil"
```

```python
start = 0
end   = 5000
step  = 500
```

```python
if not databricks:
    data_dir = "../data"
else:
    data_dir = "/dbfs/mnt/group01"

shape_map_path = os.path.join(data_dir,  "nyc/zone")
soa_shape_map = gpd.read_file(shape_map_path)
soa_shape_map_geo = soa_shape_map.to_crs("EPSG:4326")
soa_shape_map_geo["trip_number"] = np.random.randint(start, end, size=len(soa_shape_map_geo))
```

```python
soa_shape_map_geo
```

```python
def income_binning(return_colors=False):
    """Return bins and color scheme for relative median income"""
    # First bin is -10 000..-8 000, next is -8 000..-6 000, ..., final is 14 000-16 000
    bins = np.arange(start, end, step)
    cmap = plt.cm.get_cmap('Reds', len(bins))
    if not return_colors:
        return bins, cmap
    else:
        colors = []
        for i in range(cmap.N):
            rgb = cmap(i)[:3] # will return rgba, we take only first 3 so we get rgb
            colors.append(matplotlib.colors.rgb2hex(rgb))
        return bins, colors
```

```python
bins, cmap = income_binning()
colors = []
    
for i, row in soa_shape_map_geo.iterrows():
        index = bins.searchsorted(row['trip_number'])
        colors.append(matplotlib.colors.rgb2hex(cmap(index)[:3]))
```

```python
colors
```

```python
soa_shape_map_geo["color"] = colors
```
