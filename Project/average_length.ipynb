```python
import pandas as pd
import geopy.distance
import numpy as np
import matplotlib.pyplot as plt
```

```python
drivers = pd.read_csv("FOIL_2010_UNZIPPED.nosync/2010.2.csv", nrows=100)
drivers = drivers.loc[(drivers['income_per_hour'] > 10.0) & (drivers['income_per_hour'] < 80.0)]
```

```python
def pandas_dataset(file_path):
    df = pd.read_csv(file_path)
    
    df = df.loc[(df['pickup_latitude'] > 38.0) & (df['pickup_latitude'] < 42.0) & (df['pickup_longitude'] > -75.0) & (df['pickup_longitude'] < -72.0)]
    df = df.loc[(df['dropoff_latitude'] > 38.0) & (df['dropoff_latitude'] < 42.0) & (df['dropoff_longitude'] > -75.0) & (df['dropoff_longitude'] < -72.0)]
    df = df[['hack_license', 'pickup_datetime','pickup_longitude','pickup_latitude', "dropoff_longitude", "dropoff_latitude"]]
    return df
df = pandas_dataset('FOIL_2010_UNZIPPED.nosync/trip_data_2.csv')
```

```python
distance_dict = {}

for i in range(10, 80):
    distance_dict[i] = []

for i, j in zip(drivers["income_per_hour"], drivers["hack_license"]):
    if i < 80:
        new_df = df.loc[df["hack_license"] == j]
        distance = 0
        for a, b, c, d in zip(new_df['pickup_longitude'], new_df['pickup_latitude'], new_df['dropoff_longitude'], new_df['dropoff_latitude']):
            distance += geopy.distance.distance((a,b), (c, d)).km
            print(geopy.distance.distance((a,b), (c, d)).km)
        distance_dict[int(np.round(i))].append(distance/len(new_df))
    
```

```python
x_list = []
y_list = []

for i in range(10, 64):
    if len(distance_dict[i]) > 30:
        x_list.append(i)
        y_list.append(np.mean(distance_dict[i]))
```

```python
plt.plot(x_list, y_list)
plt.xlabel("Average dollar per hour")
plt.ylabel("Average length of trip (km)")
plt.show()
```
